## If Statements with User Defined Functions

```python
def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-Range Product'
    elif (row['prices'] >5) and (row['prices'] <=15): 
        return 'Mid-Range Product'
    elif row['prices'] >15:
        return 'High-Range Product'
    else: return 'Not Enough Data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

## Questions 2 & 3

```python
# create a list assigning values to each day

results = []

for value in orders_prods_merged['orders_day_of_week']:
    if value in(0,1):
        results.append('Busiest Days')
    elif value in(3,4):
        results.append('Slowest Days')
    else:
        results.append('Regular Days')
```

```python
# data check 

results
```

```python
# add a new column in the data frame 

orders_prods_merged['busiest_days'] = results
```

```python
# check frequencies of the new column in the data frame

orders_prods_merged['busiest_days'].value_counts(dropna = False)
```

The function behaved as expected. The value count in "slowest days" is equal to the sum the values for days 3 and 4 that we found above. 

## Questions 4 & 5

```python
# find the frequencies for each hour of the day

orders_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

The 6 busiest hours are: 10, 11, 14, 15, 13, 12
The 6 slowest hours are: 3, 4, 2, 5, 1, 0

```python
# create a list assigning values to each hour

results1 = []

for value in orders_prods_merged['order_hour_of_day']:
    if value in(10,11,14,15,13,12):
        results1.append('Most Orders')
    elif value in(3,4,2,5,1,0):
        results1.append('Fewest Orders')
    else:
        results1.append('Average Orders')
```

```python
# data check

results1
```

```python
# create a new column

orders_prods_merged['busiest_period_of_day'] = results1
```

```python
# check frequencies in the new column

orders_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
