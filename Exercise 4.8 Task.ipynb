# Question 4

```python
# create a new max order column by grouping by user id and find the max value of order number

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# data check to ensure new column is present

ords_prods_merge.head()
```

```python
# create flags for each category of customer

ords_prods_merge.loc[ords_prods_merge['max_order'] >40, 'loyalty_flag'] ='Loyal Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <=10, 'loyalty_flag'] ='New Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <=40) & (ords_prods_merge['max_order'] >10),
                     'loyalty_flag']= "Regular Customer"
```

```python
# data check for frequencies of each new flag 

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

# Question 5

```python
# group data by the loyalty flag column and return desciptive statistics for the prices column after grouping

ords_prods_merge.groupby('loyalty_flag')['prices'].describe()
```

The average prices for the different customer types are contained here within the mean column. This shows loyal customers have the lowest order price, new customers have the highest, and regular customers fall in the middle

# Question 6

```python
# create a average price column by grouping by user id and then finding the mean of the prices column after grouping

ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# data check to ensure the new column is in place

ords_prods_merge.head()
```

```python
# create flags for high and low spenders 

ords_prods_merge.loc[ords_prods_merge['average_price'] >=10, 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] <10, 'spending_flag'] = 'Low Spender'
```

```python
# data check for frequency of each flag in the new column

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

# Question 7

```python
ords_prods_merge['days_since_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# data check to ensure the new column is in place

ords_prods_merge.head()
```

```python
# create flags for frequency buckets

ords_prods_merge.loc[ords_prods_merge['days_since_median'] >20, 'frequency_flag'] = 'Non Frequent Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['days_since_median'] <=10, 'frequency_flag'] = 'Frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['days_since_median'] >10) & (ords_prods_merge['days_since_median'] <=20), 
                    'frequency_flag'] = 'Regular Customer'
```

```python
# data check for frequency of each flag in the new column

ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
